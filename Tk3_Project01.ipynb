```python
!pip install sodapy

import pandas as pd
from sodapy import Socrata
from datetime import datetime, timedelta

client = Socrata("data.ny.gov", None)
results = client.get("wujg-7c2s", limit=1500000)

results_df = pd.DataFrame.from_records(results)
res=pd.read_csv('https://data.ny.gov/resource/wujg-7c2s.csv?$query=SELECT%20transit_timestamp%2C%20transit_mode%2C%20station_complex_id%2C%20station_complex%2C%20borough%2C%20payment_method%2C%20fare_class_category%2C%20ridership%2C%20transfers%2C%20latitude%2C%20longitude%2C%20georeference%20WHERE%20%60transit_timestamp%60%20%3E%3D%20%20AND%20%60transit_timestamp%60%20%3C%20%20ORDER%20BY%20transit_timestamp%20ASC')
```
